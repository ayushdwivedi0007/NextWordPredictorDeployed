```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Define the list of characters to remove
characters_to_remove = [ ',', '?', '!', '-', '_', '"', "'", '(', ')']

# Open the file in read mode and read its content into a variable
with open('data.txt', 'r', encoding='utf-8') as file:
    faqs = file.read()



```

```python
for char in characters_to_remove:
    faqs = faqs.replace(char, '')
print(faqs)
```

```python
tokenizer = Tokenizer(num_words=5000)  # Limit to 5000 most frequent words

```

```python
tokenizer.fit_on_texts([faqs])
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
for sentence in faqs.split('.'):
    print(sentence)
```

```python
cleaned_faqs = ' '.join(faqs.split())

print(cleaned_faqs)
```

```python
input_sequences = []
for sentence in faqs.split('.'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(tokenized_sentence)):
        input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
from tensorflow.keras.models import Model

# Define input shape and embedding size
maxlen = X.shape[1]
max_features = 8428
embed_size = 100

# Input layer
inp = Input(shape=(maxlen, ))

# Embedding layer
x = Embedding(input_dim=max_features, output_dim=embed_size)(inp)

# LSTM layer
x = LSTM(150)(x)

# Add a fully connected layer
x = Dense(8428, activation="softmax")(x)

# Define the model
model = Model(inputs=inp, outputs=x)

```

```python
# Compile the model with sparse_categorical_crossentropy
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=60, batch_size=64)
```

```python
model.summary()
```

```python
from tensorflow.keras.models import load_model
import numpy as np
import pickle
import time
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the saved tokenizer FIRST
with open('tokenizer.pkl', 'rb') as file:
    tokenizer = pickle.load(file)

# Then load the trained model
model = load_model('next_word_predictor_model.keras')

text = " u are sure "

# The rest of your prediction code
for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=98, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
